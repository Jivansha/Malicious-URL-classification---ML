```python
from urllib.parse import urlparse
```

```python
def https_pr(url):
    urldis = urlparse(url)
    if urldis.scheme == 'https':
        return 1
    else :
        return 0
```

```python
db['https_present']= db['url'].apply(https_pr)
```

```python
o = urlparse('http://mol.com-ho.me/cv_itworx.doc')
o
```

```python
def httpadd(url):
    urldis = urlparse(url)
    if urldis.scheme == '':
        return ('http://' + url)
    else:
        return url
    
db['url'] = db['url'].apply(httpadd)
```

```python
def pridomlen(url):
    urldis = urlparse(url)
    return len(urldis.netloc)
```

```python
db['primDom_len']= db['url'].apply(pridomlen)
```

```python
def pathlen(url):
    urldis = urlparse(url)
    return len(urldis.path)
```

```python
db['path_len'] = db['url'].apply(pathlen)
```

```python
def port_pr(url):
    urldis = urlparse(url)
    if isinstance(urldis.port,int):
        return 1
    else:
        return 0
```
